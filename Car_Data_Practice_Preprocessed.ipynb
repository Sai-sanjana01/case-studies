## 1) Instruction(For data cleaning)
##### + Find all null values in the dataset. if there is any null value in any columns then fill it with the mean of the column .

```python
df.isnull().sum()
```

```python
df.notnull().sum()
```

```python
df.notna().sum()
```

```python
df.describe(include = "all")
```

```python
df["Cylinders"].fillna(df["Cylinders"].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
df["Make"] = df["Make"].fillna("Unknown")
```

```python
df["Model"] = df["Model"].fillna("Unknown")
```

```python
df[["Type","Origin","DriveTrain","MSRP","Invoice"]]= df[["Type","Origin","DriveTrain","MSRP","Invoice"]].fillna("Unkown")
```

```python
columns_to_fill = ["EngineSize","Horsepower","MPG_City","MPG_Highway","Weight","Wheelbase","Length"]
```

```python
df[columns_to_fill] = df[columns_to_fill].fillna(df[columns_to_fill].mean())
```
